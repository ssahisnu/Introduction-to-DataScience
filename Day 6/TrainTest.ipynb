```python
import numpy as np
from pylab import *
from matplotlib import pyplot as plt
from sklearn.metrics import r2_score
```

```python
np.random.seed(2)
pageSpeeds = np.random.normal(3.0,1.0,1000)
purchaseAmount = np.random.normal(50,10,1000) / pageSpeeds
scatter(pageSpeeds,purchaseAmount)
```

Since the data is randomly generated we straight away split the first 80% of the data inorder to train and the rest to test.
For most real case scenarios, the training data has to be picked randomly. Shuffling can be done using np.random.shuffle()

```python
trainX = pageSpeeds[:800]
testX = pageSpeeds[800:]

trainY = purchaseAmount[:800]
testY = purchaseAmount[800:]
```

```python
scatter(trainX,trainY)
```

```python
scatter(testX,testY)
```

```python
x = np.array(trainX)
y = np.array(trainY)

p = np.poly1d(np.polyfit(x,y,4))
```

```python
xp = np.linspace(0,7,100)
# axes = plt.axes()
# axes.set_xlim([0,7])
# axes.set_ylim([0, 200])
plt.scatter(x,y)
plt.plot(xp,p(xp),'r')
plt.show()
```

```python
testx = np.array(testX)
testy = np.array(testY)

p = np.poly1d(np.polyfit(x,y,4))
```

```python
# axes = plt.axes()
# axes.set_xlim([0,7])
# axes.set_ylim([0, 200])
plt.scatter(testx,testy)
plt.plot(xp,p(xp),'r')
plt.show()
```

```python
r2 = r2_score(testy,p(testx));
r2
```

```python
r2 = r2_score(y,p(x))
r2
```
