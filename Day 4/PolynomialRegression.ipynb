# Polynomial Regression

```python
import numpy as np
from matplotlib import pyplot as plt
from scipy import stats
from sklearn.metrics import r2_score
```

```python
pageSpeeds = np.random.normal(5.0,1.0,1000)
purchaseAmounts = np.random.normal(200.0,10.0,1000) / pageSpeeds
plt.scatter(pageSpeeds,purchaseAmounts)
```

```python
x = np.array(pageSpeeds)
y = np.array(purchaseAmounts)
p =np.poly1d(np.polyfit(x,y,3))
```

```python
xp = np.linspace(0,7,5)
plt.scatter(x,y)
plt.plot(xp,p(xp),'r')
```

### R-Square Value

```python
r2 = r2_score(y,p(x))
r2
```
