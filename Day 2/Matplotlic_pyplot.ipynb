```python
from matplotlib import pyplot as plt
import numpy as np

%matplotlib inline
```

```python
x = np.linspace(0,10,100)
y = x**2
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y,'r')
axes.set_xlabel('x')
axes.set_xlabel('y')
axes.set_title('y = x^2')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8]) #main axes
axes2 = fig.add_axes([0.2,0.5,0.4,0.3]) #inset axes

#main figure
axes1.plot(x,y,'b')
axes1.set_xlabel('x')
axes1.set_xlabel('y')
axes1.set_title('y = x^2')

#insert
axes2.plot(y,x,'g')
axes2.set_xlabel('y')
axes2.set_xlabel('x')
axes2.set_title('y = x^2')
```

```python
fig.savefig('plot.png')
```

```python
fig = plt.figure()
ax = fig.add_axes([0.1,0.1,0.8,0.8])
ax.plot(x,x**2,label = " y = x**2 ")
ax.plot(x,x**3,label = " y = x**3 ")
ax.legend(loc=2) #upper left corner; 1- top right corner; 3 - bottom left; 4 - bottom right...
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Comparison')
```
